```python
from IPython.display import Audio
from typing import cast
```

```python
from dataset_loader import LibriSpeech, LibriSpeechDataset, Dataset
```

```python
loader = LibriSpeech()
```

```python
loader.download()
```

```python
loader.prepare(verbose=True)
```

```python
dataset = loader.test_clean()
```

```python
d = dataset.to_dict()
dataset = LibriSpeechDataset.from_dict(d)

p = dataset.__getstate__()
dataset = LibriSpeechDataset.__setstate__(p)

p = dataset.__getstate__()
dataset = cast(LibriSpeechDataset, Dataset.__setstate__(p))
```

```python
id_set = set[str]()
for sample in dataset:
    # print(f"ID: {sample.id}")
    if sample.id in id_set:
        raise ValueError(f"Duplicate recording ID found: {sample.id}")
    id_set.add(sample.id)

len(dataset)
```

```python
dataset_10to20 = dataset.slice(10, 20)
dataset_30to40 = dataset.slice(30, 40)

keyset_10to20 = set[str]()
keyset_30to40 = set[str]()

keyset_10to20_for_ori = set[str]()
keyset_30to40_for_ori = set[str]()

for sample in dataset_10to20:
    keyset_10to20.add(sample.id)
for sample in dataset_30to40:
    keyset_30to40.add(sample.id)
for idx, sample in enumerate(dataset):
    if 10 <= idx < 20:
        keyset_10to20_for_ori.add(sample.id)
    if 30 <= idx < 40:
        keyset_30to40_for_ori.add(sample.id)

if keyset_10to20 != keyset_10to20_for_ori:
    raise ValueError("Slicing 10 to 20 keys do not match!")
if keyset_30to40 != keyset_30to40_for_ori:
    raise ValueError("Slicing 30 to 40 keys do not match!")
```

```python
dataset_10to20_and_30to40 = dataset_10to20 + dataset_30to40
keyset_10to20_and_30to40 = set[str]()

for sample in dataset_10to20_and_30to40:
    keyset_10to20_and_30to40.add(sample.id)

if keyset_10to20_and_30to40 != (keyset_10to20_for_ori | keyset_30to40_for_ori):
    raise ValueError("Combined slicing keys do not match!")
```

```python
from dataset_loader import ASRDataset
from dataset_loader.wrapper import DatasetWrapper
```

```python
asr_dataset = ASRDataset(dataset=dataset)
```

```python
d = asr_dataset.to_dict()
asr_dataset = ASRDataset[str, None].from_dict(d)

p = asr_dataset.__getstate__()
asr_dataset = cast(ASRDataset[str, None], ASRDataset.__setstate__(p))

p = asr_dataset.__getstate__()
asr_dataset = cast(ASRDataset[str, None], DatasetWrapper.__setstate__(p))
```

```python
asr_id_set = set[str]()
for sample in asr_dataset:
    # print(f"{sample.id}")
    asr_id_set.add(sample.id)

if len(asr_id_set) != len(asr_dataset):
    raise ValueError("Duplicate recording IDs found!")
len(asr_dataset)
```

```python
asr_dataset_10to20 = asr_dataset.slice(10, 20)
asr_dataset_30to40 = asr_dataset.slice(30, 40)

asr_keyset_10to20 = set[str]()
asr_keyset_30to40 = set[str]()

asr_keyset_10to20_for_ori = set[str]()
asr_keyset_30to40_for_ori = set[str]()

for sample in asr_dataset_10to20:
    asr_keyset_10to20.add(sample.id)
for sample in asr_dataset_30to40:
    asr_keyset_30to40.add(sample.id)
for idx, sample in enumerate(asr_dataset):
    if 10 <= idx < 20:
        asr_keyset_10to20_for_ori.add(sample.id)
    if 30 <= idx < 40:
        asr_keyset_30to40_for_ori.add(sample.id)

if asr_keyset_10to20 != asr_keyset_10to20_for_ori:
    raise ValueError("Slicing 10 to 20 keys do not match!")
if asr_keyset_30to40 != asr_keyset_30to40_for_ori:
    raise ValueError("Slicing 30 to 40 keys do not match!")
```

```python
asr_dataset_10to20_and_30to40 = asr_dataset_10to20 + asr_dataset_30to40
asr_keyset_10to20_and_30to40 = set[str]()

for asr_sample in asr_dataset_10to20_and_30to40:
    asr_keyset_10to20_and_30to40.add(asr_sample.id)

if asr_keyset_10to20_and_30to40 != (asr_keyset_10to20_for_ori | asr_keyset_30to40_for_ori):
    raise ValueError("Combined slicing keys do not match!")
```
