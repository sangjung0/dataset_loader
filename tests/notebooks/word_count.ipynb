```python
from dataset_loader.esic import ESICv1
from dataset_loader.tedlium import Tedlium
from dataset_loader.ksponspeech import KSPonSpeech
from dataset_loader.librispeech import LibriSpeech
from dataset_loader.zerothkorean import ZerothKorean
```

```python
def count_word(dataset, anchor:int = -1):
    all = 0
    i = 0
    for i, sample in enumerate(dataset):
        all += len(sample.data["ref"].split())
        if all >= anchor and anchor != -1:
            break
    return all, i
```

```python
anchor = 40213
```

```python
esic_dataset = ESICv1().dev()
print(f"ESIC: {count_word(esic_dataset, anchor)}")
del esic_dataset
```

```python
zeroth_korean_dataset = ZerothKorean().train().sample(-1)
print(f"Zeroth Korean: {count_word(zeroth_korean_dataset, anchor)}")
del zeroth_korean_dataset
```

```python
tedlium_dataset = Tedlium().train().sample(-1)
print(f"Tedlium: {count_word(tedlium_dataset, anchor)}")
del tedlium_dataset
```

```python
libri_clean_dataset = LibriSpeech().train_clean_360().sample(-1)
print(f"Libri Clean: {count_word(libri_clean_dataset, anchor)}")
del libri_clean_dataset
```
